```python
import PyPDF2                             #复制页面
pdf1File = open('meetingminutes.pdf','rb')
pdf2File = open('meetingminutes2.pdf','rb')
pdf1reader = PyPDF2.PdfReader(pdf1File)
pdf2reader = PyPDF2.PdfReader(pdf2File)
pdfWriter = PyPDF2.PdfWriter()

for pageNum in range(len(pdf1reader.pages)):
    pageObj = pdf1reader.pages[pageNum]
    pdfWriter.add_page(pageObj)

for pageNum in range(len(pdf2reader.pages)):
    pageObj = pdf2reader.pages[pageNum]
    pdfWriter.add_page(pageObj)
    
pdfOutputFile = open('combinedminutes.pdf','wb')
pdfWriter.write(pdfOutputFile)
pdfOutputFile.close()
pdf1File.close()
pdf2File.close()
```

```python
import PyPDF2
minutesFile = open('meetingminutes.pdf','rb')
reader = PyPDF2.PdfReader(minutesFile)
minutesFirstPage = reader.pages[0]
waterreader = PyPDF2.PdfReader(open('watermark.pdf','rb'))
minutesFirstPage.merge_page(waterreader.pages[0])
pdfWriter = PyPDF2.PdfWriter()
pdfWriter.add_page(minutesFirstPage)

for pageNum in range(1,len(reader.pages)):
    pageObj = reader.pages[pageNum]
    pdfWriter.add_page(pageObj)


resultPdfFile=open('watermarkedCover.pdf','wb')
pdfWriter.write(resultPdfFile)
minutesFile.close()
resultPdfFile.close()
```

```python
import PyPDF2                                  #加密PDF
pdfFile = open('meetingminutes.pdf','rb')
pdfReader = PyPDF2.PdfFileReader(pdfFile)
pdfWriter = PyPDF2.PdfFileWriter()
for pageNum in range(pdfReader.numPages):
    pdfWriter.addPage(pdfReader.getPage(pageNum))
    
pdfWriter.encrypt('swordfish')
resultPdf=open('encryptedminutes.pdf','wb')
pdfWriter.write(resultPdf)
resultPdf.close()
```

```python
import PyPDF2, os                   #记得要把加密pdf删了，再ctrl+z回收，这个程序可以跑

# Get all the PDF filenames.
pdfFiles = []
for filename in os.listdir('.'):
    if filename.endswith('.pdf'):
        pdfFiles.append(filename)
pdfFiles.sort(key=str.lower)

pdfWriter = PyPDF2.PdfFileWriter()

# Loop through all the PDF files.
for filename in pdfFiles:
    pdfFileObj = open(filename, 'rb')
    pdfReader = PyPDF2.PdfFileReader(pdfFileObj)

    # Loop through all the pages (except the first) and add them.
    for pageNum in range(1, pdfReader.numPages):
        pageObj = pdfReader.getPage(pageNum)
        pdfWriter.addPage(pageObj)

# Save the resulting PDF to a file.
pdfOutput = open('allminutes.pdf', 'wb')
pdfWriter.write(pdfOutput)
pdfOutput.close()
```

```python
import docx

def getText(filename):
    doc = docx.Document(filename)
    fullText = []
    for para in doc.paragraphs:
        fullText.append(' '+para.text) #可加缩进
    return '\n'.join(fullText)  #段落之间增加空行+\n

print(getText('demo.docx'))
```
